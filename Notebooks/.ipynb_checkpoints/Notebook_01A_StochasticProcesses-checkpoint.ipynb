```python
import numpy as np
import matplotlib.pyplot as plt
```

## Wiener Process

```python
T = 5.0
n_steps = 1000
dt = T/n_steps
W = np.zeros(n_steps)
times = np.linspace(0, T, n_steps)
```

```python
n_paths = 20
for p in range(1, n_paths):
    Z = np.random.normal(loc=0.0, scale=1.0, size=n_steps)
    W[0] = 0
    for i in range(1, n_steps):
        W[i] = W[i-1] + Z[i] * np.sqrt(dt)
    plt.plot(times, W)
plt.xlabel('Time')
plt.ylabel('$W_t$')
plt.savefig('brownian_motion.png')
```

## Adding Volatility and Linear Drift

```python
T = 5.0
mu = 0.1
sigma = 0.20

n_steps = 1000
dt = T/n_steps
sqdt = np.sqrt(dt)
```

```python
Z = np.random.normal(loc=0.0, scale=1.0, size=n_steps)
x = np.zeros(n_steps)
times = np.linspace(0, T, n_steps)
for i in range(1, n_steps):
    x[i] = x[i-1] + mu * dt + sigma * Z[i-1] * sqdt
    
plt.plot(times,x);
```

## Lognormal Stock Price Process

```python
S = np.zeros(n_steps)
t = np.zeros(n_steps)
S[0] = 100.0
mu = 0.05
sigma = 0.10
```

```python
Z = np.random.normal(loc=0.0, scale=1.0, size=n_steps)

for i in range(1, n_steps):    
    S[i] = S[i-1] * (1.0 + mu * dt + Z[i] * sigma * sqdt)
    t[i] = i*dt

plt.figure(figsize=(6, 4), dpi=100)
plt.xlabel("Time(T)")
plt.ylabel("S(T)")
plt.plot(t,S, 'b')
```

## The Terminal Stock Price Distribution

```python
n_paths = 20000
T = 5.0
S0 = 100.0
mu = 0.05
sigma = 0.10
```

```python
S = np.zeros(n_paths)
Z = np.random.normal(loc=0.0, scale=1.0, size=n_paths)

for p in range(0, n_paths):
    S[p] = S0 * np.exp( (mu-sigma*sigma/2) * T + sigma * Z[p] * np.sqrt(T)) 
```

```python
plt.hist(S, bins=100);
plt.xlabel("S(T)")
plt.ylabel("Frequency")
```

```python
# Define the range for x values
smin = np.min(S)
smax = np.max(S)

ST = np.linspace(smin, smax, 100)  # Generates values from 0.01 to 3

# Function to calculate the lognormal PDF
def lognormal_pdf(tau, s0, sT, mu, sigma):
    Z = (np.log(sT/s0) - (mu - sigma*sigma/2)*tau)/sigma/np.sqrt(tau)
    pdf = 1.0/ np.sqrt(2 * np.pi) * np.exp(-Z*Z/2.0)
    ds = 1.0 / (sigma * np.sqrt(tau) * sT + 1e-10)
    pdf = pdf * ds
    return pdf

# Calculate the PDF for each x value using Vectorization
pdf = lognormal_pdf(T, S0, ST, mu, sigma)
```

```python
# I convert the frequency plot to a density plot 
plt.hist(S, 100, density=True, label='density');
plt.plot(ST, pdf, label='pdf');
plt.xlabel("S(T)")
plt.ylabel("Density")
plt.legend();
```
